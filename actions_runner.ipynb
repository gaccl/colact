```python
import os
# Set User and Groups to Add
os.environ['RUNNER_USER'] = 'runner'
os.environ['SUDO_GROUP'] = 'sudo'
```

```python
# Set NOPASSWD to /etc/sudoers
!sed -i 's/\ /\t/g' /etc/sudoers
!sed -i 's/ALL\:ALL/ALL/g' /etc/sudoers
!sed -i 's/ALL=(ALL)\tALL/ALL=(ALL)\tNOPASSWD\:\tALL/g' /etc/sudoers
!sed -i 's/\t/\ /g' /etc/sudoers
```

```python
# Set Path Environments
os.environ['RUNNER_HOME'] = '/'
if os.environ['RUNNER_USER'] != 'root':
  os.environ['RUNNER_HOME'] += 'home/'
os.environ['RUNNER_HOME'] += os.environ['RUNNER_USER']
# Github Actions Runner Directory
os.environ['RUNNER_PATH'] = os.environ['RUNNER_HOME'] + '/actions-runner'
```

```python
# Create Directory
import os
# Safe to mkdir(path) & chdir(path) & chwon(path,user)
def mkdir(path, user):
  if os.path.exists(path):
    print(' '.join([
      path,
      'is',
      'already',
      'exists.'
      'Skipping',
      'mkdir'
    ]))
  else:
    print(' '.join([
      'Create',
      'Directory',
      path
    ]))
    os.mkdir(path)
  os.system(' '.join([
    'chown',
    '-R'
    ':'.join([
      user,
      user
    ]),
    path
  ]))
  os.chdir(path)
# Create Directory ${RUNNER_HOME}
mkdir(os.environ['RUNNER_HOME'], os.environ['RUNNER_USER'])
# Create Directory ${RUNNER_PATH}
mkdir(os.environ['RUNNER_PATH'], os.environ['RUNNER_USER'])
```

```python
import os
# Set Github Actions Runner Release Download Options
os.environ['RUNNER_OS'] = 'linux'
os.environ['RUNNER_ARCH'] = 'x64'
os.environ['RUNNER_VERSION'] = '2.288.1'
```

```python
import os
# THIS ALLOWS RUN AS ROOT OR IN SUDO
os.environ['RUNNER_ALLOW_RUNASROOT'] = "1" # THIS ENV WILL DO
# Set Github Actions Runner Config Environments
os.environ['RUNNER_PROTOCOL'] = 'https:'
os.environ['RUNNER_HOST'] = 'github.com'
os.environ['RUNNER_ORG'] = '<ORGANIZATION_NAME>'
os.environ['RUNNER_REPO'] = '<REPOSITORY_NAME>'
os.environ['RUNNER_TOKEN'] = '<RUNNER_TOKEN>'
```

```python
import os
# Set Command Environments
os.environ['RUNAS'] = 'sudo -u '
```

```python
# Add ${RUNNER_USER}
!useradd "${RUNNER_USER}"
```

```python
# Add ${RUNNER_USER} to ${SUDO_GROUP}
!usermod -aG "${SUDO_GROUP}" "${RUNNER_USER}" || true
```

```python
# Create HOME
!mkdir -p "${RUNNER_HOME}" || true
# Change ${RUNNER_HOME} 's Own to ${RUNNER_USER}:${RUNNER_USER}
!chown -R \
  "${RUNNER_USER}":"${RUNNER_USER}" \
  "${RUNNER_HOME}"
```

```python
# Update packages
!${RUNAS} ${RUNNER_USER} sudo apt-get update -y
```

```python
import os
# Change Working Directory to ${RUNNER_HOME}
os.chdir(os.environ['RUNNER_HOME'])
# Create Directory ${RUNNER_PATH} as ${RUNNER_USER}
!${RUNAS} ${RUNNER_USER} mkdir -p ${RUNNER_PATH}
```

```python
import os
# Change Working Directory to ${RUNNER_PATH}
os.chdir(os.environ['RUNNER_PATH'])
# Download Github Actions Runner Release 
!${RUNAS} ${RUNNER_USER} curl \
    -o actions-runner-${RUNNER_OS}-${RUNNER_ARCH}-${RUNNER_VERSION}.tar.gz \
    -L https://github.com/actions/runner/releases/download/v${RUNNER_VERSION}/actions-runner-${RUNNER_OS}-${RUNNER_ARCH}-${RUNNER_VERSION}.tar.gz
```

```python
import os
# Change Working Directory to ${RUNNER_PATH}
os.chdir(os.environ['RUNNER_PATH'])
# Extract Github Actions Runner Release 
!${RUNAS} ${RUNNER_USER} tar -xzvf "actions-runner-${RUNNER_OS}-${RUNNER_ARCH}-${RUNNER_VERSION}.tar.gz"
```

```python
import os
# Change Working Directory to ${RUNNER_PATH}
os.chdir(os.environ['RUNNER_PATH'])
# Config Github Actions Runner
!export RUNNER_ALLOW_RUNASROOT="1"; ./config.sh \
    --url "${RUNNER_PROTOCOL}//${RUNNER_HOST}/${RUNNER_ORG}/${RUNNER_REPO}" \
    --token "${RUNNER_TOKEN}"
!ls -la
```

```python
import os
# Change Working Directory to ${RUNNER_PATH}
os.chdir(os.environ['RUNNER_PATH'])
# Run Github Actions Runner
!./run.sh
```
